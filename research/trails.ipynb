```python
from langchain.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
#Extract data from the PDF file
def load_pdf_file(dataset):
    loader = DirectoryLoader(dataset, glob="*.pdf", loader_cls=PyPDFLoader)
    documents = loader.load()
    return documents
```

```python
extracted_data = load_pdf_file(dataset ='dataset/')
```

```python
extracted_data 
```

```python
#split the data into text chunks 
def text_split(extracted_data):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=20)
    text_chunks = text_splitter.split_documents(extracted_data)
    return text_chunks
```

```python
text_chunks = text_split(extracted_data)
print(f"Total number of text chunks: {len(text_chunks)}")
```

```python
import os
from dotenv import load_dotenv
import pinecone
from pinecone import ServerlessSpec

# Load environment variables
load_dotenv()

# Fetch the API key
api_key = os.getenv("PINECONE_API_KEY")

# ✅ Confirm it's loaded (optional debug)
assert api_key is not None, "PINECONE_API_KEY not loaded!"

# Initialize Pinecone
pc = pinecone.Pinecone(api_key=api_key)

# Define and create the index
index_name = "medicalbot"

pc.create_index(
    name=index_name,
    dimension=384,
    metric="cosine",
    spec=ServerlessSpec(
        cloud="aws",
        region="us-east-1"
    )
)
```

```python
import os 
os.environ["PINECONE_API_KEY"] = api_key
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
```

```python
#Embed each chunk and upsert the embeddings into the Pinecone index
from langchain_pinecone import PineconeVectorStore

docsearch = PineconeVectorStore.from_documents(
    documents=text_chunks,
    index_name=index_name,
    embedding=embeddings,
)
```

```python
#Load existing index
from langchain_pinecone import PineconeVectorStore
docsearch = PineconeVectorStore.from_existing_index(
    index_name=index_name,
    embedding=embeddings,
)
```

```python
docsearch
```

```python
retriver = docsearch.as_retriever(search_type="similarity", search_kwargs={"k": 3})
```

```python
retriver_docs = retriver.invoke("What is Acne?")
retriver_docs
```

```python
from langchain_openai import OpenAI
llm = OpenAI(temperature=0.4, max_tokens=500)
```

```python
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate

system_prompt = (

"You are an assistant that helps people find information about medical conditions and treatments"
"Use the following pieces of retrieved context to answer"
"the question. If you don't know the answer, just say that you don't know"
"Use three sentences maximum and keep the answer concise"
"{context}"
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        ("human", "{input}"),
    ]
)
```

```python
question_answer_chain = create_stuff_documents_chain(llm, prompt)
rag_chain = create_retrieval_chain(retriver, question_answer_chain)
```

```python
response = rag_chain.invoke({"input": "What is Acne?"})
print(response["answer"])
```

```python
response = rag_chain.invoke({"input": "What is stats?"})
print(response["answer"])
```
